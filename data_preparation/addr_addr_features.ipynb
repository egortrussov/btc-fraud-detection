```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set(style="darkgrid")

np.random.seed(0)
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

```python
addrTx = pd.read_csv("../dataset/Elliptic++ Dataset/AddrTx_edgelist.csv")
txAddr = pd.read_csv("../dataset/Elliptic++ Dataset/TxAddr_edgelist.csv")
tx_features = pd.read_csv("../dataset/Elliptic++ Dataset/txs_features.csv")
```

```python
tx_features_small = tx_features[[
    "txId",
    "in_BTC_total",
    "out_BTC_total",
    "Time step",
]]
```

```python
addr_addr_tx = addrTx \
    .merge(
        txAddr,
        how="left",
        on="txId"
    ) \
    .merge(
        tx_features_small,
        how="left",
        on="txId"
    )

addr_addr_tx.info()
```

```python
addr_addr_aggregated = addr_addr_tx \
    .groupby(["input_address", "output_address"]) \
    .agg({
        "in_BTC_total": "sum",
        "out_BTC_total": "sum",
        "Time step": "nunique",
        "txId": "nunique",
    }) \
    .reset_index() \
    .rename(columns={"Time step": "active_ts_cnt", "txId": "tx_cnt"})
```

```python
addr_addr_aggregated.to_csv("../dataset/custom/AddrAddr_EdgeFeatures.csv", index=False)
```
