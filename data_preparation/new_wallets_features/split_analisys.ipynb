```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set(style="darkgrid")

np.random.seed(0)
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

```python
train_data_raw = pd.read_csv("../../dataset/custom/features_fixed/wallets_features__train.csv")
test_data_raw = pd.read_csv("../../dataset/custom/features_fixed/wallets_features__test.csv")
```

```python
train_data_raw_licit = train_data_raw[train_data_raw["class"] != 1]
train_data_raw_illicit = train_data_raw[train_data_raw["class"] == 1]

test_data_raw_licit = test_data_raw[test_data_raw["class"] != 1]
test_data_raw_illicit = test_data_raw[test_data_raw["class"] == 1]
```

```python
def plot_metrics(licit, illicit, cutoff, ax, title=""):
    sns.histplot(
        licit[licit <= cutoff],
        ax=ax,
        bins=50,
        stat="probability"
    )
    ax.set_title(title)

    sns.histplot(
        illicit[illicit <= cutoff],
        ax=ax,
        bins=50,
        stat="probability"
    )
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))

plot_metrics(
    train_data_raw["incoming_tx_cnt"],
    test_data_raw["incoming_tx_cnt"],
    cutoff=20,
    ax=ax[0]
)
plot_metrics(
    train_data_raw["outcoming_tx_cnt"],
    test_data_raw["outcoming_tx_cnt"],
    cutoff=20,
    ax=ax[1]
)
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))

plot_metrics(
    train_data_raw["btc_sent_total"],
    test_data_raw["btc_sent_total"],
    cutoff=20,
    ax=ax[0]
)
plot_metrics(
    train_data_raw["btc_received_total"],
    test_data_raw["btc_received_total"],
    cutoff=20,
    ax=ax[1]
)
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))

plot_metrics(
    train_data_raw["incoming_tx_fees_min"],
    test_data_raw["incoming_tx_fees_min"],
    cutoff=20,
    ax=ax[0]
)
plot_metrics(
    train_data_raw["outcoming_tx_fees_min"],
    test_data_raw["outcoming_tx_fees_min"],
    cutoff=20,
    ax=ax[1]
)
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))

plot_metrics(
    train_data_raw["output_address_cnt"],
    test_data_raw["output_address_cnt"],
    cutoff=20,
    ax=ax[0]
)
plot_metrics(
    train_data_raw["input_address_cnt"],
    test_data_raw["input_address_cnt"],
    cutoff=20,
    ax=ax[1]
)
```

# Licit-illicit on 2 time serieses

```python
fig, ax = plt.subplots(10,2, figsize=(10,60))

plot_metrics(
    train_data_raw_licit["btc_sent_min"],
    test_data_raw_licit["btc_sent_min"],
    cutoff=20,
    ax=ax[0][0],
    title="btc_sent_min (licit)"
)
plot_metrics(
    train_data_raw_illicit["btc_sent_min"],
    test_data_raw_illicit["btc_sent_min"],
    cutoff=20,
    ax=ax[0][1],
    title="btc_sent_min (illicit)"
)

plot_metrics(
    train_data_raw_licit["btc_received_min"],
    test_data_raw_licit["btc_received_min"],
    cutoff=20,
    ax=ax[1][0],
    title="btc_received_min (licit)"
)
plot_metrics(
    train_data_raw_illicit["btc_received_min"],
    test_data_raw_illicit["btc_received_min"],
    cutoff=20,
    ax=ax[1][1],
    title="btc_received_min (illicit)"
)


plot_metrics(
    train_data_raw_licit["btc_received_total"],
    test_data_raw_licit["btc_received_total"],
    cutoff=20,
    ax=ax[2][0],
    title="btc_received_total (licit)"
)
plot_metrics(
    train_data_raw_illicit["btc_received_total"],
    test_data_raw_illicit["btc_received_total"],
    cutoff=20,
    ax=ax[2][1],
    title="btc_received_total (illicit)"
)


plot_metrics(
    train_data_raw_licit["outcoming_tx_cnt"],
    test_data_raw_licit["outcoming_tx_cnt"],
    cutoff=20,
    ax=ax[3][0],
    title="outcoming_tx_cnt (licit)"
)
plot_metrics(
    train_data_raw_illicit["outcoming_tx_cnt"],
    test_data_raw_illicit["outcoming_tx_cnt"],
    cutoff=20,
    ax=ax[3][1],
    title="outcoming_tx_cnt (illicit)"
)

plot_metrics(
    train_data_raw_licit["user_addr_cnt"],
    test_data_raw_licit["user_addr_cnt"],
    cutoff=20,
    ax=ax[4][0],
    title="user_addr_cnt (licit)"
)
plot_metrics(
    train_data_raw_illicit["user_addr_cnt"],
    test_data_raw_illicit["user_addr_cnt"],
    cutoff=20,
    ax=ax[4][1],
    title="user_addr_cnt (illicit)"
)
```
